```python
import pandas as pd
import numpy as np
import re
import plotly_express as px
```

```python
f = open("WhatsApp Chat with GDSC Team 2022-23.txt","r",encoding="utf-8")
```

```python
data = f.read()
print(data)
```

##### To seperate messages and dates from data

```python
pattern =  "\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{1,2}\s"
```

```python
messages = re.split(pattern,data)[1:]
```

```python
messages
```

```python
dates = re.findall(pattern , data)
dates
```

```python
df = pd.DataFrame({"Dates":dates,"messages":messages})
df
```

#### Removing time from dates column

```python
def date(a:str):
    return a[0:10]
```

```python
df["Dates"]= df["Dates"].apply(date)
```

```python
df
```

#### Removing "pm -" from messages

```python
def remove_pm(a:str):
    return a[5:]
```

```python
df["messages"] = df["messages"].apply(remove_pm)
```

```python
df = df.iloc[3:]
df['messages']
```

#### Creating Sender_name column 

```python
def extract_name_no(text_data:str):
    # Define a regex pattern to match names
    name_pattern = re.compile(r'^([^:\n]+):')

    # Extract names from each line
    names = [match.group(1) for line in text_data.splitlines() if (match := name_pattern.search(line))]

    return names
```

```python
df["Sender_name"] = df["messages"].apply(extract_name_no)
```

```python
df
```

#### Removing Name and Number from start of messages

```python
messages = "+91 958226: *Online meet today at 7 pm ISTsdddddddddddbfb"

def removeSendernamefromMesaages(messages:str):
    pattern = re.compile(r'^([^:]+):')

    match = pattern.match(messages)
    if match:
        sender_name = match.group(1)
        result = messages[len(sender_name)+1:]
        return result
    else:
        return "Sender's Name not found"
    
removeSendernamefromMesaages(messages)
```

```python
df["messages"] = df["messages"].apply(removeSendernamefromMesaages)
```

```python
df
```

### Removing "\n" in messages column

```python
def remove(text:str):
    pattern = "\n"
    if pattern in text:
        text = re.sub(pattern, "", text)
    return text

df["messages"] = df["messages"].apply(remove)
df
```

```python
df["Sender_name"] =df["Sender_name"].astype(str)
```

```python
df["Sender_name"] = df["Sender_name"].apply(lambda x: x.replace(x,x[2:-2]))
df
```

```python
phone_number_mapping = {
    '+91 96532 96937': 'Dhiraj Chauhan',
    '+91 79778 83880': 'Maitri',
    '+91 93722 00217': 'Sanskar Mishra',
    '+91 70216 67972': 'Vanshika',
    '+91 82915 24320': 'Jayesh',
    '+91 98704 54408': 'Ishir Agarwaal',
    '+91 99875 68422': 'Prashant'
}

# Replace phone numbers with names using the mapping
df['Sender_name'] = df['Sender_name'].replace(phone_number_mapping)
```
